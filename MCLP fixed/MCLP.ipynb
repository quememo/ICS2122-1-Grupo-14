```python
from mclp import *
```

```python
%matplotlib inline
```

```python
import pandas as pd 
import numpy as np
```

### Generate input data

```python
bases = pd.read_csv('Datos/bases.csv', delimiter=';')
```

```python
llamadas = pd.read_csv('Datos/eventos.csv', delimiter=';')
```

```python
llamadas = llamadas[["COORDENADA_X", "COORDENADA_Y"]]
llamadas.head()
```

```python
llamadas.describe()
```

```python
# Dataframe to list
nodos_bases = bases.values.tolist()
nodos_llamadas = llamadas.values.tolist()
```

```python
# converting list to array
llamadas_array = np.array(nodos_llamadas)
print(llamadas_array)
print(type(llamadas_array))
```

```python
# converting list to array
bases_array = np.array(nodos_bases)
print(type(bases_array))
```

```python
# Plot input data
plot_input(llamadas_array)
```

```python
# Plot input data
plot_input(bases_array)
```

### Run the optimization

#### Radio = 18

```python
# Maximum Coverage Location Problem
# Service radius of each site !
radius = 18
# Number of sites to select
K = 20
# Candidate site size (random sites generated)
M = 26

# Run mclp opt_sites is the location of optimal sites and f is the points covered
opt_sites,f = mclp(llamadas_array,K,radius,M,bases_array)

# Plot the result
plot_result(llamadas_array,opt_sites,radius)
```

```python
plot_result2(llamadas_array,opt_sites,radius)
```

```python
# Bases seleccionadas 
print(opt_sites)
```
